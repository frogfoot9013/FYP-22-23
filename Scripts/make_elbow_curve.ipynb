```python
import matplotlib.pyplot as plt
import pandas as pd
import os
core_dir = os.getcwd()

df_1 = pd.read_csv(core_dir+"/../Scores/elbow_graph_kmeans.csv")
df_2 = pd.read_csv(core_dir+"/../Scores/elbow_graph_bisecting_kmeans.csv")

max_inertia = df_2.loc[0,"Inertia"]

fig, ax = plt.subplots()

plt.xlim([100, 600])
plt.ylim([0, max_inertia])
plt.ylabel("Inertia")
plt.xlabel("Clusters")
ax.plot(df_1.loc[:,"Clusters"], df_1.loc[:,"Inertia"], **{'color': 'lightsteelblue', 'marker': 'o'})
ax.plot(df_2.loc[:,"Clusters"], df_2.loc[:,"Inertia"], **{'color': 'red', 'marker': 'o'})
plt.legend(labels=["K-Means", "Bisecting K-Means"], loc=4, borderpad=1)
plt.show()
```
